```python
# å®‰è£ï¼š pip install googlemaps
import googlemaps
import os

API_KEY = os.getenv("GOOGLE_MAPS_API_KEY")  # å»ºè­°å¾ç’°å¢ƒè®Šæ•¸è®€å–
gmaps = googlemaps.Client(key=API_KEY)

# 1. è¨­å®šæœå°‹ä½ç½®ï¼ˆç¯„ä¾‹ï¼šå°åŒ—å¸‚ä¸­å¿ƒï¼‰
lat, lng = 25.0330, 121.5654
radius = 1000  # å…¬å°º

# 2. Nearby Search å‘¼å«
results = gmaps.places_nearby(
    location=(lat, lng),
    radius=radius,
    type='restaurant',  # å°ˆæ‰¾é¤å»³
    open_now=False      # å¯é¸ï¼šåªæ‰¾ç›®å‰é–‹æ”¾
)

for place in results.get('results', []):
    name = place.get('name')
    address = place.get('vicinity')
    place_id = place.get('place_id')
    rating = place.get('rating')
    print(f"Name: {name}, Address: {address}, Rating: {rating}, Place ID: {place_id}")

    # è‹¥æƒ³æ‹¿æ›´å¤šç´°ç¯€ï¼Œå¯å‘¼å« Place Details
    details = gmaps.place(
        place_id=place_id,
        fields=['name','formatted_address','formatted_phone_number','website','rating','user_ratings_total']
    )
    deta = details.get('result', {})
    print("   Phone:", deta.get('formatted_phone_number'))
    print("   Website:", deta.get('website'))
    print("   Total Ratings:", deta.get('user_ratings_total'))
```

```python
import pandas as pd
import numpy as np
import random
import gradio as gr
import asyncio
import json

# --- 2. æ ¸å¿ƒåŠŸèƒ½å‡½æ•¸ ---

def filter_and_recommend(df, price_filter, distance_filter, is_available):
    """æ ¹æ“šä½¿ç”¨è€…è¼¸å…¥ç¯©é¸ä¸¦éš¨æ©Ÿæ¨è–¦ 3 å®¶é¤å»³"""

    # è½‰æ›ç¯©é¸æ¢ä»¶ç‚ºåˆ—è¡¨
    price_list = price_filter if price_filter else df['åƒ¹ä½'].unique()
    distance_list = distance_filter if distance_filter else df['è·é›¢'].unique()
    # Gradio CheckboxGroup å›å‚³åˆ—è¡¨ï¼Œä½†å–®ä¸€å‹¾é¸æ™‚éœ€è™•ç†
    available_status = ['æ˜¯'] if is_available else ['æ˜¯', 'å¦']

    # æ‡‰ç”¨ç¯©é¸æ¢ä»¶
    filtered_df = df[
        df['åƒ¹ä½'].isin(price_list) &
        df['è·é›¢'].isin(distance_list) &
        df['æ˜¯å¦ç‡Ÿæ¥­'].isin(available_status)
    ].copy()

    if filtered_df.empty:
        return pd.DataFrame(), "æ²’æœ‰æ‰¾åˆ°ç¬¦åˆæ¢ä»¶çš„é¤å»³ã€‚", ""

    # éš¨æ©ŸæŠ½å– 3 å®¶ï¼Œä¸è¶³ 3 å®¶å‰‡å…¨æ•¸å–å‡º
    n_recommend = min(3, len(filtered_df))
    recommended_df = filtered_df.sample(n=n_recommend, replace=False)

    # æº–å‚™æ¨è–¦æ¸…å–®æ–‡å­—
    recommend_list = "\n".join([
        f"{i+1}. {row['é¤å»³åç¨±']} ({row['åƒ¹ä½']} / {row['è·é›¢']}) - {row['ç‰¹è‰²æè¿°']}"
        for i, row in recommended_df.iterrows()
    ])

    # æº–å‚™ AI æ‘˜è¦çš„è¼¸å…¥å­—ä¸²
    ai_input = "ã€".join([
        f"{row['é¤å»³åç¨±']}ï¼Œåƒ¹ä½{row['åƒ¹ä½']}ï¼Œè·é›¢{row['è·é›¢']}ï¼Œç‰¹è‰²ï¼š{row['ç‰¹è‰²æè¿°']}"
        for _, row in recommended_df.iterrows()
    ])

    return recommended_df, recommend_list, ai_input


async def generate_summary(ai_input):
    """å‘¼å« LLM ç”¢ç”Ÿæ±ºç­–å»ºè­°æ‘˜è¦ (ä½¿ç”¨ gemini-2.5-pro)"""
    if not ai_input:
        return "è«‹å…ˆé»æ“Šã€Œé–‹å§‹æ¨è–¦ã€ä»¥ç”¢ç”Ÿé¤å»³æ¸…å–®ã€‚"

    # Fallback if model configuration failed
    await asyncio.sleep(1) # æ¨¡æ“¬ç¶²è·¯å»¶é²
    mock_summary = f"""
    **AI åŠ©ç† (æ¨¡æ“¬):** ä»Šå¤©çš„é¸æ“‡æ˜¯ï¼š{'ã€'.join([r.split('(')[0].strip() for r in ai_input.split('ã€')])}ã€‚
    å¦‚æœæ‚¨è¿½æ±‚**æ•ˆç‡å’ŒéŠ…æ¿åƒ¹**ï¼Œ{ai_input.split('ã€')[0].split('ï¼Œ')[0]} (åƒ¹ä½{ai_input.split('ã€')[0].split('ï¼Œ')[1].split('åƒ¹ä½')[1]}) è‚¯å®šæ˜¯é¦–é¸ï¼Œå¿«é€Ÿåˆé£½è¶³ã€‚
    è‹¥æƒ³ä¾†é»**å„€å¼æ„Ÿå’Œé«˜å“è³ª**ï¼Œ{'ã€'.join(ai_input.split('ã€')[1:])} æä¾›äº†æ›´å¥½çš„ç”¨é¤é«”é©—ï¼Œä½†è·é›¢å’ŒèŠ±è²»æœƒç¨é«˜ä¸€é»ã€‚
    æ±ºå®šæ¬Šåœ¨ä½ ï¼
    (è«‹ç¢ºä¿æ‚¨å·²å®‰è£ 'google-genai' ä¸¦åœ¨ Colab Secrets ä¸­è¨­ç½® 'gemini' API key ä»¥å•Ÿç”¨çœŸå¯¦ AI æ¨è–¦)
    """
    return mock_summary.strip()


    system_prompt = (
        "æ‚¨æ˜¯ä¸€ä½å¹½é»˜ä¸”æ¥µå…·æ•ˆç‡çš„åˆé¤æ±ºç­–é¡§å•ã€‚æ‚¨çš„ä»»å‹™æ˜¯æ ¹æ“šæä¾›çš„é¤å»³æ¸…å–®ï¼Œ"
        "åˆ†æå…¶å„ªç¼ºé»ï¼ˆå¦‚åƒ¹æ ¼ã€è·é›¢ã€ç‰¹è‰²ï¼‰ä¸¦ç¸½çµï¼Œå¹«åŠ©ä½¿ç”¨è€…åœ¨ 3 å®¶ä¸­å¿«é€Ÿåšå‡ºæ±ºå®šã€‚"
        "è«‹ä»¥è¼•é¬†å£å»ï¼Œç”¨ **ç¹é«”ä¸­æ–‡** æ’°å¯«ä¸€å€‹ç´„ **100 å­—** çš„æ±ºç­–å»ºè­°æ‘˜è¦ï¼Œä¸è¶…é 150 å­—ã€‚"
        "è«‹é¿å…ä½¿ç”¨ Markdown æ¨™é¡Œå’Œç·¨è™Ÿæ¸…å–®ã€‚"
    )

    user_query = f"é€™æ˜¯ä¸‰å®¶æ¨è–¦çš„é¤å»³æ¸…å–®ï¼š{ai_input}ã€‚è«‹ç¶œåˆè€ƒé‡å…¶åƒ¹æ ¼ã€è·é›¢å’Œç‰¹è‰²ï¼Œç‚ºæˆ‘æä¾›ä¸€ä»½å¿«é€Ÿæ±ºç­–å»ºè­°ã€‚"

    try:
        # ä½¿ç”¨ asyncio.to_thread è®“åŒæ­¥çš„ SDK å‘¼å«ä¸æœƒé˜»å¡ Gradio çš„äº‹ä»¶è¿´åœˆ
        response = await asyncio.to_thread(
            model.generate_content,
            contents=user_query,
            config=genai.types.GenerateContentConfig(
                system_instruction=system_prompt,
                temperature=0.7,
            )
        )
        return response.text.strip()

    except Exception as e:
        error_msg = f"å‘¼å« Gemini API å¤±æ•—ï¼Œè«‹æª¢æŸ¥ API é‡‘é‘°æˆ–é…é¡ï¼š{e}ã€‚"
        return error_msg


# ä»‹é¢æ›´æ–°å‡½æ•¸
async def wrapper_recommend(price_filter, distance_filter, is_available):
    """Gradio ä¸»è¦æµç¨‹ï¼šç¯©é¸ -> æ¨è–¦ -> å‘¼å« AI æ‘˜è¦"""

    # 1. ç¯©é¸èˆ‡æ¨è–¦
    recommended_df, recommend_list, ai_input = filter_and_recommend(
        df_restaurants, price_filter, distance_filter, is_available
    )

    # 2. å‘¼å« AI æ‘˜è¦
    ai_summary = await generate_summary(ai_input)

    # 3. æº–å‚™çµ¦ Gradio è¼¸å‡º
    # å°‡æ¨è–¦çµæœè½‰æ›ç‚º Dataframe (åƒ…é¡¯ç¤ºåç¨±ã€åƒ¹ä½ã€è·é›¢)
    display_df = recommended_df[['é¤å»³åç¨±', 'åƒ¹ä½', 'è·é›¢']].reset_index(drop=True)

    # è™•ç†ç©ºçµæœ
    if display_df.empty:
        return pd.DataFrame(), "æ²’æœ‰æ‰¾åˆ°ç¬¦åˆæ¢ä»¶çš„é¤å»³ã€‚", "ï¼ˆç„¡å»ºè­°ï¼‰"

    return display_df, recommend_list, ai_summary


# --- 3. Gradio ä»‹é¢é…ç½® ---

# è¨­å®š Gradio ä»‹é¢å…ƒä»¶
with gr.Blocks(title="è¿·ä½ é»‘å®¢æ¾åˆé¤æ±ºç­–å™¨") as demo:
    gr.Markdown(
        """
        # ğŸ½ï¸ è¿·ä½ é»‘å®¢æ¾åˆé¤/èšæœƒæ±ºç­–å™¨
        ### ç›®æ¨™ï¼šå¾é¤å»³æ¸…å–®ä¸­ç¯©é¸ï¼Œä¸¦éš¨æ©Ÿæ¨è–¦ 3 å®¶ï¼Œæœ€å¾Œç”± AI åŠ©æ‚¨å¿«é€Ÿæ±ºç­–ï¼
        **ğŸš¨ æ³¨æ„ï¼šè«‹ç¢ºä¿æ‚¨å·²åœ¨ Colab ä¸­è®€å– Google Sheet æ•¸æ“šï¼Œå¦å‰‡å°‡ä½¿ç”¨æ¨¡æ“¬æ•¸æ“šã€‚**
        """
    )

    with gr.Row():
        with gr.Column(scale=1):
            gr.Markdown("## æ¢ä»¶ç¯©é¸")
            price_select = gr.CheckboxGroup(
                ['$', '$$', '$$$'],
                label="åƒ¹ä½éæ¿¾ (å¯è¤‡é¸)",
                value=['$', '$$'],
                info="($: éŠ…æ¿/150å…§, $$: 150-300, $$$: 300+)"
            )
            distance_select = gr.CheckboxGroup(
                ['çŸ­', 'ä¸­', 'é•·'],
                label="è·é›¢éæ¿¾ (å¯è¤‡é¸)",
                value=['çŸ­', 'ä¸­'],
                info="(çŸ­: æ­¥è¡Œ5åˆ†, ä¸­: æ­¥è¡Œ10åˆ†, é•·: éœ€é¨è»Š)"
            )
            available_toggle = gr.Checkbox(
                label="åªçœ‹ä»Šæ—¥ç‡Ÿæ¥­",
                value=True
            )
            recommend_btn = gr.Button("ğŸš€ é–‹å§‹æ¨è–¦")

        with gr.Column(scale=2):
            gr.Markdown("## æ¨è–¦æ¸…å–® (éš¨æ©Ÿ 3 å®¶)")
            output_dataframe = gr.DataFrame(
                headers=["é¤å»³åç¨±", "åƒ¹ä½", "è·é›¢"],
                datatype=["str", "str", "str"],
                label="éš¨æ©Ÿæ¨è–¦çµæœ"
            )
            output_list = gr.Textbox(label="æ¨è–¦æ¸…å–®ç´°ç¯€")

    gr.Markdown("## ğŸ¤– AI æ±ºç­–å»ºè­° (ç´„ 100 å­—)")
    output_summary = gr.Textbox(
        label="AI æ±ºç­–é¡§å•å ±å‘Š",
        interactive=False,
        info="AI å°‡æ ¹æ“šæ¨è–¦çµæœï¼Œç¸½çµå„ªç¼ºé»ï¼Œå¹«åŠ©æ‚¨å¿«é€Ÿåšå‡ºé¸æ“‡ã€‚"
    )

    # å®šç¾©é»æ“ŠæŒ‰éˆ•å¾Œçš„è¡Œç‚º (ä½¿ç”¨ç•°æ­¥åŒ…è£å‡½æ•¸)
    recommend_btn.click(
        fn=wrapper_recommend,
        inputs=[
            price_select,
            distance_select,
            available_toggle
        ],
        outputs=[output_dataframe, output_list, output_summary]
    )

    gr.Markdown(
        """
        ---
        **[é–‹ç™¼è€…å‚™è¨»]** AI æ‘˜è¦å·²è¨­å®šç‚ºä½¿ç”¨ `gemini-2.5-pro` å‘¼å«ã€‚è«‹ç¢ºä¿æ‚¨å·²åŸ·è¡Œä»¥ä¸‹æ­¥é©Ÿï¼š
        1.  åœ¨ Colab ä¸­å®‰è£ `google-genai` ç­‰å¿…è¦å¥—ä»¶ã€‚
        2.  åœ¨ Colab Secrets ä¸­è¨­å®šåç‚º `gemini` çš„ API é‡‘é‘°ã€‚
        """
    )


# é‹è¡Œ Gradio æ‡‰ç”¨
if __name__ == "__main__":
    # åœ¨ Colab æˆ–ä¸€èˆ¬ç’°å¢ƒä¸­é‹è¡Œ
    print("Gradio æ‡‰ç”¨æ­£åœ¨å•Ÿå‹•...")
    demo.launch(inbrowser=True)
```
